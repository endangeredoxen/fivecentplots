# <b>Gantt Charts<b>

<b><font color="blue" style="font-family:'Courier New'">fivecentplots </font></b> can be used to make basic Gantt charts for project management using  `fcp.gantt`.  At a minimum, the `gantt` function requires the following keywords:
<ul>
<li>`df`: a pandas DataFrame</li>
<li>`x`: a list containing the names of two columns in the DataFrame, representing the "Start" and "Stop" datetime values
<li>`y`: the name of the DataFrame column containing the tasks for the Gantt chart</li>
</ul>

## Setup

### Imports

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import fivecentplots as fcp
import pandas as pd
import numpy as np
import os, sys, pdb
osjoin = os.path.join
db = pdb.set_trace
```

### Sample data

Read some fake bar chart data

```python
df = pd.read_csv(osjoin(os.path.dirname(fcp.__file__), 'tests', 'fake_data_gantt.csv'))
df.head()
```

### Set theme

Optionally set the design theme

```python
#fcp.set_theme('gray')
#fcp.set_theme('white')
```

### Other

```python
SHOW = False
```

## Basic chart

A basic Gantt chart showing tasks across a time axis is created as follows:

```python
fcp.gantt(df=df, x=['Start', 'Stop'], y='Task', ax_size=[600, 400], show=SHOW)
```

As with other plots, tick label fonts, sizes, and rotations can all be controlled using the correct ``kwarg``.  See <a href="ticks.html">Grids and ticks</a> for more details.

## Legends

Legends can be used to further distinguish various tasks by assignee or other factors.  If the same task has multiple assignees, they will be split into two different bars in the chart.

```python
fcp.gantt(df=df, x=['Start', 'Stop'], y='Task', legend='Assigned', ax_size=[600, 400], show=SHOW)
```
