# <b>pie<b>

<b><font color="blue" style="font-family:'Courier New'">fivecentplots </font></b> can plot pie charts using  `fcp.pie`.  At a minimum, the `bar` function requires the following keywords:
<ul>
<li> df: a pandas DataFrame</li>
<li> x: the name of the categorical data column (i.e., the labels)
<li> y: the name of the DataFrame column containing the values for the data in column x</li>
</ul>

## Setup

### Imports

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import fivecentplots as fcp
import pandas as pd
import numpy as np
import os, sys, pdb
osjoin = os.path.join
db = pdb.set_trace
```

### Sample data

Read some fake bar chart data

```python
df = pd.read_csv(osjoin(os.path.dirname(fcp.__file__), 'tests', 'fake_data_bar.csv'))
df.loc[df.pH < 0, 'pH'] = -df.pH
df.head()
```

### Set theme

Optionally set the design theme

```python
#fcp.set_theme('gray')
#fcp.set_theme('white')
```

### Other

```python
SHOW = False
```

## Basic Plot

Consider a DataFrame that contains several sets of measured pH data for various liquids.  First we plot only one set of measured data:

```python
fcp.pie(df, x='Liquid', y='pH', show=SHOW, filter='Measurement=="A" & T [C]==25', startangle=90, alpha=0.85)
```

## Donut

We can turn our pie chart into a donut chart by using the keyword `innerradius` (aka the `width` attribute in MPL's `wedgeprops`).  The inner radius should be lower than the main `radius` parameter (set to 1 by default).  In this example, we also shift the position of the percentages slightly to look cleaner.

```python
fcp.pie(df, x='Liquid', y='pH', show=SHOW, filter='Measurement=="A" & T [C]==25', startangle=90, alpha=0.85, innerradius=0.5, pctdistance=0.75)
```

## Grouping

Like the <font style="font-family:'Courier New'">plot</font> function, pie charts can grouped in various ways to provide further insight into the data set.  This includes legends, `row` and/or `col` subplot values and `wrap` values.  

### Legend

```python
fcp.pie(df, x='Liquid', y='pH', show=SHOW, filter='Measurement=="A" & T [C]==25', startangle=90, alpha=0.85, legend=True)
```

### Row/column plot

We can separate different conditions using row/col plots as shown below:

```python
fcp.pie(df, x='Liquid', y='pH', show=SHOW, col='Measurement', row='T [C]', legend=True, ax_size=[250, 250])
```

### Wrap plot

Alternatively, we can use a `wrap` plot to visualize:

```python
fcp.pie(df, x='Liquid', y='pH', show=SHOW, wrap='Measurement', ax_size=[250, 250], legend=True)
```

## Options

### Labeled percents

```python
fcp.pie(df, x='Liquid', y='pH', show=SHOW, filter='Measurement=="A" & T [C]==25', startangle=90, alpha=0.85, percents=True)
```

### Explode

```python
fcp.pie(df, x='Liquid', y='pH', show=SHOW, filter='Measurement=="A" & T [C]==25', explode=(0,0.1), startangle=90, alpha=0.85, percents=True)
```

### Shadow

```python
fcp.pie(df, x='Liquid', y='pH', show=SHOW, filter='Measurement=="A" & T [C]==25', explode=(0,0.1), shadow=True, startangle=90, percents=True)
```

### Start angle

```python
fcp.pie(df, x='Liquid', y='pH', show=SHOW, filter='Measurement=="A" & T [C]==25', explode=(0,0.1), startangle=0, percents=True)
```
